```python
# Importing libraries
import numpy as np
import pandas as pd
```

```python
# pesticides data

pest_data = pd.read_csv('/content/FAOSTAT_pesticide.csv')
pest_data
```

```python
# remove unnecessary columns

pest_data = pest_data.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code', 'Element', 'Item Code', 'Item', 'Year Code', 'Flag', 'Flag Description'], axis = 1)
pest_data
```

```python
# crop data

crop_data = pd.read_csv('/content/FAOSTAT_crops.csv')
crop_data
```

```python
# remove unnecessary columns

crop_data = crop_data.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code', 'Element', 'Item Code (CPC)', 'Year Code', 'Unit', 'Flag', 'Flag Description', 'Note'], axis = 1)
crop_data
```

```python
# rainfall data

rain_data = pd.read_csv('/content/Updated_rainfall.csv')
rain_data
```

```python
# average rainfall per year

# Group by 'Year' and calculate the mean for each year
rain_data = rain_data.groupby('Year')['Rainfall (mm)'].mean().reset_index()
rain_data
```

```python
# temperature data

temp_data = pd.read_csv('/content/temp_csv.csv')
temp_data
```

```python
# remove unnecessary rows so data is only until 2021

index_dropped = [12, 13]

temp_data = temp_data.drop(index_dropped)
temp_data
```

```python
# Merging data

pest_data.rename({'Area': 'Country', 'Value': 'Pesticides (Tons)'}, axis = 1, inplace = True)
crop_data.rename({'Area': 'Country', 'Value': 'Yield (hg/ha)'}, axis = 1, inplace = True)
temp_data.rename({'Area': 'Country', 'avg_temp': 'Temperature (Celcius)'}, axis = 1, inplace = True)
```

```python
rain_temp = pd.merge(rain_data, temp_data, on=['Year'])
rain_temp_crop = pd.merge(rain_temp, crop_data, on=['Year'])
rain_temp_crop_pest = pd.merge(rain_temp_crop, pest_data, on=['Year', 'Country'])
rain_temp_crop_pest.drop(['Unit'], axis = 1, inplace = True)
merged_data = rain_temp_crop_pest[['Year', 'Country', 'Item', 'Rainfall (mm)', 'Temperature (Celcius)', 'Yield (hg/ha)', 'Pesticides (Tons)']]
merged_data
```

```python
# Save the merged DataFrame to a CSV file
merged_data.to_csv('Final_merged_data.csv', index=False)
```
